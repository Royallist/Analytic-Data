# Analisis dan Perhitungan

# 1
untuk mendapatkan persebaran penduduk di Indonesia, kita harus mendeskripsikan terlebih dahulu kolom baru, yaitu total seluruh penduduk (Laki-laki + Perempuan). kemudian kita akan bentuk choropleth untuk melihat sebaran data pada provinsi di indonesia

```python
# Library used
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import matplotlib.patches as mpatches
import seaborn as sns

#Membuat kolom Total Penduduk
data_podes['Total Penduduk']= data_podes['Jumlah Penduduk Laki-laki'] + data_podes['Jumlah Penduduk Perempuan']

#Membuat data berdasarkan provinsi
data_podes_provinsi = data_podes['Total Penduduk'].groupby(data_podes['NAMA_PROV']).sum()

# Import GeoJSON Data
path = 'https://raw.githubusercontent.com/bimaputra1/School_Partitipation_Rates_with_GeoPandas/master/gadm36_IDN_1.json'
df_geo = gpd.read_file(path)

```

```python
#Membuat dataframe baru guna di join kan dengan data GeoJSON untuk pembuatan choropleth
data_podes_provinsi=data_podes['Total Penduduk'].groupby(data_podes['NAMA_PROV']).sum()
datageo_podes=pd.DataFrame()
datageo_podes['Total Penduduk']=data_podes_provinsi.values
datageo_podes['Provinsi']=df_geo['NAME_1']
datageo_podes

#melakukan join dataframe
df_join = df_geo.merge(datageo_podes, how='inner', left_on="NAME_1", right_on="Provinsi")
df_join = df_join[['Provinsi','Total Penduduk','geometry']]
df_join.head()
```

```python
# Mendeskripsikan jangkauan data dari plot yang akan dibuat
vmin, vmax = df_join['Total Penduduk'].min(),df_join['Total Penduduk'].max()
# menentukan figur dan sumbu dari plot
fig, ax = plt.subplots(1, figsize=(30, 10))
# remove the axis
ax.axis('off')
# menambahkan judul plot
title = 'Persebaran penduduk di-Indonesia'
ax.set_title(title, fontdict={'fontsize': '25', 'fontweight' : '3'})
# membuat catatan kaki
ax.annotate('Source: Badan Pusat Statistik Indonesia',xy=(0.1, .08),  xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', fontsize=12 ,color='#555555')
# mendeskripsikan warna dan nilai pada keterangan plot
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))

# membuat plot
df_join.plot(column=datageo_podes['Total Penduduk'], cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', norm=plt.Normalize(vmin=vmin, vmax=vmax))

# menambahkan nama provinsi
df_join['coords'] = df_join['geometry'].apply(lambda x: x.representative_point().coords[:])
df_join['coords'] = [coords[0] for coords in df_join['coords']]
for idx, row in df_join.iterrows():
    plt.annotate(text=row['Provinsi'], xy=row['coords'], horizontalalignment='center')

# membuat keterangan pada posisi sesuai sumbu
cbar = fig.colorbar(sm, ax=ax)
```

### Intepretasi
Dari choroplath diatas, dapat dilihat bahwa penduduk indonesia banyak berada pada provinsi dipulau jawa. range antara penduduk yang berada dijawa dan di luar jawa memiliki nilai yang sangat tinggi, ini bisa dilihat bahwa jumlah penduduk paling sedikit dipulau jawa memiliki jumlah penduduk yang lebih tinggi dari rata-rata penduduk di kalimantan. sehingga bisa dikatan bahwa penyebaran penduduk indonesia berdasarkan data podes 2011 mengkrucut pada 1 titik saja yaitu tepatnya di provinsi yang ada di pulau jawa.

Untuk menjawab pertanyaan ke 5, kita dapat menggunakan pengujian korelasi menggunakan uji spearman-r dikarenkan data diasumsikan memiliki outlier dan uji spearmen-r tidak sensitif terhadap outlier sehingga uji korelasi dapat maksimal

## Pengujian Hipotesis

### 1. Hipotesis

H0 = Jumlah rumah sakit dan jumlah keluarga saling bebas
H1 = Ada Hubungan antara Jumlah rumah sakit dan jumlah keluarga

### 2. Besaran yang diperlukan

alpha = 0.05

### 3. Statistik Uji


### 4. Daerah Penolakan

Tolak H0 jika P-value < alpha
Terima H0 jika P-value >= alpha

```python
from scipy import stats

# Calculate Spearman correlation
correlation, p_value = stats.spearmanr(data_podes['Jumlah Rumah Sakit'].groupby(data_podes['NAMA_PROV']).sum().values, data_podes['Jumlah Keluarga'].groupby(data_podes['NAMA_PROV']).sum().values)

# Print the Spearman correlation coefficient
print(f"Spearman-r: {correlation:.2f}", "P-Value:",p_value)
```

### 5 Kesimpulan

Dari hasil uji spearmen diatas didapat niali P-value sebesar 6.141413836123712e-11 < dari aplhoa = 0.05. ini berarti pada taraf kepercayaan 95%, kita percaya ada Hubungan antara Jumlah rumah sakit dan jumlah keluarga.
hal ini juga didukung dari nilai spearman-r nya yang hampir mendekati 1. ini berarti pembangunan rumah sakit sudah tepat karena dibangun di kawansan yang ramai penduduk sehingga dapat menjangkau banyak orang

Untuk menjawab pertanyaan no 6, kita dapat melakukan eksplorasi data terhadap jumlah kelurahan yang ada

```python
#Menghitung jumlah kelurahan berdasakan provinsi
jumlah_desa=data_podes.groupby('NAMA_PROV').size().reset_index(name='Jumlah Desa/Kelurahan')
```

```python
jumlah_desa.describe() #melihat nilai statistika deskriptif
```

```python
nilai_skew=jumlah_desa['Jumlah Desa/Kelurahan'].skew()
print('Nilai Skewness:',nilai_skew)
```

```python
nilai_kurt = jumlah_desa['Jumlah Desa/Kelurahan'].kurt()
print('Nilai Kurtosis',nilai_kurt)
```

2. Dilihat dari nilai standar deviasi yang cukup tinggi, ini berarti terdapat nilai yang jauh dari rata-ratanya. Hal ini mengindikasikan bahwa terdapat nilai outlier pada data
3. Dari nilai skewness dan kurtosis data, mengindikasikan data menyebar negative skew dengan leptokurtik.

Hal ini berarti, pesebaran jumlah kelurahan/desa di Indonesia masih belum merata. terdapat beberapa provinsi yang jumlah kelurahan/desa nya jauh dari rata-rata yang ada.
